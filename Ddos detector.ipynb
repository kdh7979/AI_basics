# ----------------------------데이터 전처리----------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_dataset.csv')
df.head()
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
df.dtypes.head(10)
```

```python
df.shape
```

```python
df_sample = df.sample(frac=0.3, random_state=54)
```

```python
df_sample.shape
```

```python
labels = ['Ddos', 'Benign'] ## 라벨 
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
pie = ax.pie(df_sample['Label'].value_counts(), ## 파이차트 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       autopct=lambda p : '{:.2f}%'.format(p), ## 퍼센티지 출력
       colors = ['purple', 'lightcoral'] ## 색상 지정
       )
plt.legend(pie[0],labels) ## 범례 표시
plt.show()
```

```python
df_sample.head()
```

```python
df_sample.isna().sum().head(40)
```

```python
df_sample.isna().sum().tail(44)
```

```python
df_sample.describe()
```

```python
df_sample1 = df_sample.dropna()
print(df_sample1.shape)
```

```python
df_sample1['Label'].value_counts()
```

```python
df_ddos = df_sample1[df_sample1['Label'] == 'ddos']
df_ddos.shape
```

```python
df_benign = df_sample1[df_sample1['Label'] == 'Benign']
df_benign.shape
```

```python
df_ddos_sample = df_ddos.sample(n=100000, random_state=54)
df_benign_sample = df_benign.sample(n=100000, random_state=54)
print(df_ddos_sample.shape)
print(df_benign_sample.shape)
```

```python
df_sampled = pd.concat([df_ddos_sample, df_benign_sample], axis=0)
df_sampled.shape
```

```python
df_sampled.head()
```

```python
df_sampled = df_sampled.sort_values(by='Timestamp')
```

```python
df_sampled.head(20)
```

```python
df_sampled["Label"].value_counts()
```

```python
df_sampled.to_csv('sampled_dataset.csv')
```

# ---------------------------------상관관계 파악---------------------------------------

```python
data = pd.read_csv('sampled_dataset.csv')
```

```python
onehot = pd.get_dummies(data['Label'])
onehot.head()
```

```python
data = data.join(onehot)
```

```python
data.head()
```

```python
data.columns
```

```python
cols = ['Flow ID', 'Src IP', 'Src Port', 'Dst IP', 'Dst Port',
       'Protocol', 'Timestamp', 'Flow Duration', 'Tot Fwd Pkts',
       'Tot Bwd Pkts', 'TotLen Fwd Pkts', 'TotLen Bwd Pkts', 'Fwd Pkt Len Max',
       'Fwd Pkt Len Min', 'Fwd Pkt Len Mean', 'Fwd Pkt Len Std',
       'Bwd Pkt Len Max', 'Bwd Pkt Len Min', 'Bwd Pkt Len Mean',
       'Bwd Pkt Len Std', 'Flow Byts/s', 'Flow Pkts/s', 'Flow IAT Mean',
       'Flow IAT Std', 'Flow IAT Max', 'Flow IAT Min', 'Fwd IAT Tot',
       'Fwd IAT Mean', 'Fwd IAT Std', 'Fwd IAT Max', 'Fwd IAT Min',
       'Bwd IAT Tot', 'Bwd IAT Mean', 'Bwd IAT Std', 'Bwd IAT Max',
       'Bwd IAT Min', 'Fwd PSH Flags', 'Bwd PSH Flags', 'Fwd URG Flags',
       'Bwd URG Flags', 'Fwd Header Len', 'Bwd Header Len', 'Fwd Pkts/s',
       'Bwd Pkts/s', 'Pkt Len Min', 'Pkt Len Max', 'Pkt Len Mean',
       'Pkt Len Std', 'Pkt Len Var', 'FIN Flag Cnt', 'SYN Flag Cnt',
       'RST Flag Cnt', 'PSH Flag Cnt', 'ACK Flag Cnt', 'URG Flag Cnt',
       'CWE Flag Count', 'ECE Flag Cnt', 'Down/Up Ratio', 'Pkt Size Avg',
       'Fwd Seg Size Avg', 'Bwd Seg Size Avg', 'Fwd Byts/b Avg',
       'Fwd Pkts/b Avg', 'Fwd Blk Rate Avg', 'Bwd Byts/b Avg',
       'Bwd Pkts/b Avg', 'Bwd Blk Rate Avg', 'Subflow Fwd Pkts',
       'Subflow Fwd Byts', 'Subflow Bwd Pkts', 'Subflow Bwd Byts',
       'Init Fwd Win Byts', 'Init Bwd Win Byts', 'Fwd Act Data Pkts',
       'Fwd Seg Size Min', 'Active Mean', 'Active Std', 'Active Max',
       'Active Min', 'Idle Mean', 'Idle Std', 'Idle Max', 'Idle Min', 'Label',
       'Benign', 'ddos']
```

```python
corr = data[cols].corr(method='pearson')
```

```python
plt.figure(figsize=(50,40))
sns.set(font_scale=1.5)
sns.heatmap(corr.values,
           cbar=True,
           annot=True,
           square=True,
           fmt='.2f',
           annot_kws={'size':10},
           yticklabels=cols,
           xticklabels=cols)
plt.tight_layout()
plt.show()
```
